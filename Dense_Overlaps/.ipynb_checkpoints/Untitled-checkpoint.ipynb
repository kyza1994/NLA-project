```python
import numpy as np
from math import exp
import networkx as nx
from matplotlib import pyplot as plt
```

# Network generator:

```python
NCommunities = 6
NNodes = 30

#Average community size
A = 6
epsilon = 0.05

#Node-Community Affilation matrix
NC = np.asarray([[0. for i in range(NCommunities)] for j in range(NNodes)])
for community in range(NCommunities):
    community_size = np.random.randint(A-2,A+3)
    for i in range(community_size):
        member = np.random.randint(0,NNodes)
        NC[member,community] = 1

#Change NC matrix into weight matrix F:
for i in range(NCommunities):
    for j in range(NNodes):
        if NC[j,i] != 0:
            NC[j,i] = 3*np.random.rand()
            
plt.matshow(NC.T, fignum=100, cmap=plt.cm.gray)
plt.show()

#Node-Node affilation matrix
NN = np.asarray([[0 for i in range(NNodes)] for j in range(NNodes)])
for i in range(NNodes):
    for j in range(NNodes):
        if i != j:
            F = sum([NC[i,k]*NC[j,k] for k in range(NCommunities)])
            measure = np.random.rand()
            if measure < 1. - exp(-F) or measure < epsilon:
                NN[i,j] = 1
                NN[j,i] = 1        
print NC.T
```

```python
G = nx.from_numpy_matrix(np.matrix(NN))
pos = nx.spring_layout(G)
labels = {}
for i in xrange(NNodes):
    labels[i] = i
    
plt.figure(figsize = (10,10))
nx.draw_networkx_nodes(G, pos, node_size=300, node_color = "r")
nx.draw_networkx_labels(G, pos, labels, font_size=10)
nx.draw_networkx_edges(G, pos, G.edges())
plt.show()
```

```python
import cvxpy as cvx

F = cvx.Variable((NNodes,NCommunities))

constraints = [F >= 0]

objective = cvx.Maximize()
```

```python
total_vol = np.sum(NN)
conductances = []

local_minimals = []

for node in range(NNodes):
    neighbours = []
    for othernode in range(NNodes):
        if othernode != node:
            if NN[node,othernode] == 1:
                neighbours.append(othernode)
    
    S = neighbours + [node]

    vol_S = sum(NN[node,:])
    for current_neighbour in neighbours:
        for i in S:
            if NN[current_neighbour,i] == 1:
                vol_S += 1
        
    not_S = []
    for i in range(NNodes):
        if i not in S:
            not_S.append(i)
            
    cut = 0
    for i in neighbours:
        for j in not_S:
            if NN[i,j] == 1:
                cut +=1

    vol_not_S = 0
    for i in not_S:
        for j in not_S:
            if NN[i,j] == 1:
                vol_not_S += 1
                
    if min(vol_S, vol_not_S) != 0:
        conductances.append(cut / float(min(vol_S, vol_not_S)))
    else:
        conductances.append(0.)
        
for node in range(NNodes):
    node_conductance = conductances[node]
    flag = 1
    for othernode in range(NNodes):
        if othernode != node:
            if NN[node,othernode] == 1:
                flag = 0
                if conductances[othernode] <= node_conductance:
                    flag = 1
                    break
    
    if flag == 0:
        local_minimals.append(node)
        
print local_minimals
```
